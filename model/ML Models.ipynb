Preparing data for ML
Creating Pipeline and applying models

```
from pyspark.ml.feature import VectorAssembler #combine multiple columns into a single one

assembler = VectorAssembler(inputCols = ['v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'v7', 'v8', 'v9', 'v10', 'v11', 'v12', 'v13', 'v14', 'v15', 'v16', 'v17', 'v18', 'v19', 'v20',
 'v21', 'v22', 'v23', 'v24', 'v25', 'v26', 'v27', 'v28', 'v29', 'v30', 'v31', 'v32', 'v33', 'v34', 'v35', 'v36', 'v37', 'v38', 'v39', 'v40', 'v41', 'v42', 'v43', 'v44',
 'v45', 'v46', 'v47', 'v48', 'v49', 'v50', 'v51', 'v52', 'v53', 'v54', 'v55', 'v56', 'v57', 'v58', 'v59', 'v60', 'v61', 'v62', 'v63', 'v64', 'v65', 'v66', 'v67', 'v68', 'v69',
 'v70', 'v71', 'v72', 'v73', 'v74', 'v75', 'v76', 'v77', 'v78', 'v79', 'v80', 'v81', 'v82', 'v83', 'v84', 'v85', 'v86', 'v87', 'v88', 'v89', 'v90', 'v91', 'v92', 'v93', 'v94',
 'v95', 'v96', 'v97', 'v98', 'v99', 'v100', 'v101', 'v102', 'v103', 'v104', 'v105', 'v106', 'v107', 'v108', 'v109', 'v110', 'v111', 'v112', 'v113', 'v114', 'v115', 'v116',
 'v117', 'v118', 'v119', 'v120', 'v121', 'v122', 'v123', 'v124', 'v125', 'v126', 'v127', 'v128', 'v129', 'v130', 'v131'], outputCol = 'features')
```

```
output = assembler.transform(df)
```

```
df2 = output.select('features', 'target')
df2.show(3)
```

```
train, test = df2.randomSplit([0.7, 0.3]) #split the data
```

```
from pyspark.ml.classification import (LogisticRegression, DecisionTreeClassifier, RandomForestClassifier, GBTClassifier)
from pyspark.ml import Pipeline

logreg = LogisticRegression(labelCol = 'target', featuresCol = 'features')
dectree = DecisionTreeClassifier(labelCol = 'target', featuresCol = 'features')
randfor = RandomForestClassifier(labelCol = 'target', featuresCol = 'features')
gbclass = GBTClassifier(labelCol = 'target', featuresCol = 'features')
```

```
#fit model
logreg_model = logreg.fit(train)
dectree_model = dectree.fit(train)
randfor_model = randfor.fit(train)
gbclass_model = gbclass.fit(train)
```

```
#transform
logreg_predictions = logreg_model.transform(test)
dectree_predictions = dectree_model.transform(test)
randfor_predictions = randfor_model.transform(test)
gbclass_predictions = gbclass_model.transform(test)
```

Evaluate models by Area under ROC and Accuracy

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator #evaluate predictions

binary_evaluator = BinaryClassificationEvaluator(labelCol = 'target')

print('Logistic Regression - Area Under ROC:' , binary_evaluator.evaluate(logreg_predictions))
```

```
print('Decision Tree - Area Under ROC:', binary_evaluator.evaluate(dectree_predictions))
```

```
print('Random Forest - Area Under ROC:' , binary_evaluator.evaluate(randfor_predictions))
```

```
print('Gradient-boosted Trees - Area Under ROC:', binary_evaluator.evaluate(gbclass_predictions))
```

```
binary_evaluator.getMetricName()
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator #Evaluate model

multi_evaluator = MulticlassClassificationEvaluator(labelCol = 'target', metricName = 'accuracy')
print('Logistic Regression:', multi_evaluator.evaluate(logreg_predictions))
```

```
print('Decision Tree Accu:', multi_evaluator.evaluate(dectree_predictions))
```
