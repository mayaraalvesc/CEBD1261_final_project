```python
# File location and type
file_location = "/FileStore/tables/train.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

#display(df)
```

```python
temp_table_name = "claim"

df.createOrReplaceTempView(temp_table_name)
```

```python
import numpy as np
import pandas as pd

pdf = df.toPandas()
```

```python
for col in pdf: #converting float into float32
    if pdf[col].dtype == 'double':
            pdf[col] = pdf[col].astype('float32')
```

```python
for col in pdf: #converting int64 into int32
    if pdf[col].dtype == 'int64':
            pdf[col] = pdf[col].astype('int32')
```
